```python
print("""[2차방정식의 실수 해 계산 프로그램]
=====================================""")

print("x에 대한 2차방정식 ax^2+bx+c = 0에서")

a, b, c = map(int, input("상수 a, b, c 값을 차례로 입력하세요: ").split(','))  # 2차 방정식에 들어갈 a,b,c 값 입력

if (a > 0):#a가 0보다 클 때 식 형태
    if (a == 1):#a가 1일 때 이차항의 계수 1 생략
        if (b > 0):#b가 0보다 클 때 식 형태
            if (b == 1):#b가 1일 때 일차항의 계수1 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: x^2 + x + {0}'.format(c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: x^2 + x')
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: x^2 + x - {0}'.format(c*(-1)))
                    
            else:#b가 1이 아닌 양의 정수 일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: x^2 + {0}x + {1}'.format(b,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: x^2 + {0}x'.format(b))

                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: x^2 + {0}x - {1}'.format(b,c*(-1)))

            
        elif (b == 0):#b가 0일 때 식 형태
            if (c > 0):#c가 0보다 클 때 식 형태
                print('입력한 이차방정식: x^2 + {0}'.format(c))
                
            elif (c == 0):#c가 0일 때 식 형태
                print('입력한 이차방정식: x^2')
                    
            else:#c가 0보다 작을 때 식 형태
                print('입력한 이차방정식: x^2 - {0}'.format(c*(-1)))
        
        else:#b가 0보다 작을 때 식 형태
            if (b == -1):#b가 -1 일때 일차항 계수는 부호만 남기고 1 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: x^2 - x + {0}'.format(c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: x^2 - x')
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: x^2 - x - {0}'.format(c*(-1)))
            else:#b가 -1이 아닌 0보다 작을 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: x^2 - {0}x + {1}'.format(b*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: x^2 - {0}x'.format(b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: x^2 - {0}x - {1}'.format(b*(-1),c*(-1)))
                   
    else:#a가 1이 아닌 다른 양의 정수
        if (b > 0):#b가 0보다 클 때 식 형태
            if (b == 1):#b가 1일 때 일차항의 계수1 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + x + {1}'.format(a,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + x'.format(a))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + x - {1}'.format(a,c*(-1)))
            else:#b가 1이 아닌 양의 정수
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + {1}x + {2}'.format(a,b,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + {1}x'.format(a,b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: {0}x^2 + {1}x - {2}'.format(a,b,c*(-1)))
        
        elif (b == 0):#b가 0일 때 식 형태
            if (c > 0):#c가 0보다 클 때 식 형태
                print('입력한 이차방정식: {0}x^2 + {1}'.format(a,c))
                
            elif (c == 0):#c가 0일 때 식 형태
                print('입력한 이차방정식: {0}x^2'.format(a))
                    
            else:#c가 0보다 작을 때 식 형태
                print('입력한 이차방정식: {0}x^2 - {1}'.format(a,c*(-1)))
        
        else:#b가 0보다 작을 때
            if (b == -1):#b가 -1 일때 일차항 계수는 부호만 남기고 1 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - x + {1}'.format(a,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - x'.format(a))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - x - {1}'.format(a,c*(-1)))
            else:#b 가 -1이 아닌 다른 음의 정수
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - {1}x + {2}'.format(a,b*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - {1}x'.format(a,b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: {0}x^2 - {1}x - {2}'.format(a,b*(-1),c*(-1))) 
                
    D = b ** 2 - 4 * a * c  # 판별식으로 근 개수 구하기
    
    root1 = (-b + (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 첫 번째 근 구하기, 중근이 나올시 이 값 적용할 것
    
    root2 = (-b - (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 두 번째 근 구하기
    
    print("---------------------------------------------------------")
    
    if (D > 0):  # 근의 개수와 근 출력, 근은 소수점 6번째 자리 까지 표현
        print("2개의 실수 해가 존재합니다: x = {0:.6f}, x = {1:.6f}".format(root1, root2))  # 판별식이 0보다 클 때 근 2개

    elif (D == 0):
        print("1개의 실수 해가 존재합니다: x = {0:.6f}".format(root1))  # 판별식이 0일 때 근 1개 (중근)

    else:
        print("실수 해가 존재하지 않습니다")  # 판별식이 0보다 작을 때 근이 존재하지 않음
    
    print("=====================================")
               
elif (a < 0):#a가 0보다 작을 때 식 형태
    if (a == -1):#a가 -1일 때 이차항의 계수는 부호만 남기고 숫자 생략
        if (b > 0):#b가 0보다 클 때 식 형태
            if (b == 1):#b가 1일 때 일차항의 계수1 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - x^2 + x + {0}'.format(c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - x^2 + x')
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - x^2 + x - {0}'.format(c*(-1)))
            else:#b가 1이 아닌 양의 정수일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - x^2 + {0}x + {1}'.format(b,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - x^2 + {0}x'.format(b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - x^2 + {0}x - {1}'.format(b,c*(-1)))
            
        elif (b == 0):#b가 0일 때 식 형태
            if (c > 0):#c가 0보다 클 때 식 형태
                print('입력한 이차방정식: - x^2 + {0}'.format(c))
                
            elif (c == 0):#c가 0일 때 식 형태
                print('입력한 이차방정식: - x^2')
                    
            else:#c가 0보다 작을 때 식 형태
                print('입력한 이차방정식: - x^2 - {0}'.format(c*(-1)))
        
        else:#b가 0보다 작을 때 식 형태
            if (b == -1):#b가 -1일때 일차항의 부호만 남기고 숫자 생략
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - x^2 - x + {0}'.format(c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - x^2 - x')
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - x^2 - x - {0}'.format(c*(-1)))
            else:#b가 -1이 아닌 다른 음의 정수일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - x^2 - {0}x + {1}'.format(b*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - x^2 - {0}x'.format(b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - x^2 - {0}x - {1}'.format(b*(-1),c*(-1)))
                   
    else:#a가 -1 이 아닌 다른 음의 정수일 때 식 형태
        if (b > 0):#b가 0보다 클 때 식 형태
            if (b == 1):#b가 1일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + x + {1}'.format(a*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + x'.format(a*(-1)))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + x - {1}'.format(a*(-1),c*(-1)))
            else:#b가 1이 아닌 양의 정수일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + {1}x + {2}'.format(a*(-1),b,c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + {1}x'.format(a*(-1),b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 + {1}x - {2}'.format(a*(-1),b,c*(-1)))
            
        elif (b == 0):#b가 0일 때 식 형태
            if (c > 0):#c가 0보다 클 때 식 형태
                print('입력한 이차방정식: - {0}x^2 + {1}'.format(a*(-1),c))
                
            elif (c == 0):#c가 0일 때 식 형태
                print('입력한 이차방정식: - {0}x^2'.format(a*(-1)))
                    
            else:#c가 0보다 작을 때 식 형태
                print('입력한 이차방정식: - {0}x^2 - {1}'.format(a*(-1),c*(-1)))
        
        else:#b가 0보다 작을 때 식 형태
            if (b == -1):#b가 -1일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - x + {1}'.format(a*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - x'.format(a*(-1)))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - x - {1}'.format(a*(-1),c*(-1)))
            else:#b가 -1이 아닌 음의 정수일 때 식 형태
                if (c > 0):#c가 0보다 클 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - {1}x + {2}'.format(a*(-1),b*(-1),c))
                
                elif (c == 0):#c가 0일 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - {1}x'.format(a*(-1),b))
                    
                else:#c가 0보다 작을 때 식 형태
                    print('입력한 이차방정식: - {0}x^2 - {1}x - {2}'.format(a*(-1),b*(-1),c*(-1)))
                    
    D = b ** 2 - 4 * a * c  # 판별식으로 근 개수 구하기
    
    root1 = (-b + (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 첫 번째 근 구하기, 중근이 나올시 이 값 적용할 것
    
    root2 = (-b - (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 두 번째 근 구하기
    
    print("---------------------------------------------------------")
    
    if (D > 0):  # 근의 개수와 근 출력, 근은 소수점 6번째 자리 까지 표현
        print("2개의 실수 해가 존재합니다: x = {0:.6f}, x = {1:.6f}".format(root1, root2))  # 판별식이 0보다 클 때 근 2개

    elif (D == 0):
        print("1개의 실수 해가 존재합니다: x = {0:.6f}".format(root1))  # 판별식이 0일 때 근 1개

    else:#
        print("실수 해가 존재하지 않습니다")  # 판별식이 0보다 작을 때 근이 존재하지 않음
    
    print("=====================================")                
    
else:#a에 0이 입력 되었을 때
    print("---------------------------------------------------------")
    print("계수 a에 0을 입력하지마세요.")
    print("=====================================")
    
    
    
    
```
