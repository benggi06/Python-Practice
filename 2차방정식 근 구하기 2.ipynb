```python
print("""[2차방정식의 실수 해 계산 프로그램]
=====================================""")

print("x에 대한 2차방정식 ax^2+bx+c = 0에서")

a, b, c = map(int, input("상수 a, b, c 값을 차례로 입력하세요: ").split(','))  # 2차 방정식에 들어갈 a,b,c 값 입력

if (a !=0):#a가 0이 아닐때
    if (a > 0):#a가 0보다 클 때
        if (a == 1):#a가 1일때 계수 생략
            if (b > 0):  # b가 0보다 클 때 방정식 모양
                if (b == 1):
                    if (c > 0):
                        print("입력한 2차 방정식: x^2 + x + {2}".format(a, b, c))  # b가 0보다 크고 c가 0보다 클 때 식 모양

                    elif (c == 0):
                        print("입력한 2차 방정식: x^2 + x".format(a, b))  # b가 0보다 크고 c가 0일 때 식 모양

                    else:
                        print("입력한 2차 방정식: x^2 + x - {2}".format(a, b, c*(-1)))  # b가 0보다 크고 c가 0보다 작을 때 식 모양
                else:
                    if (c > 0):
                        print("입력한 2차 방정식: x^2 + {1}x + {2}".format(a, b, c))  # b가 0보다 크고 c가 0보다 클 때 식 모양
                
                    elif (c == 0):
                        print("입력한 2차 방정식: x^2 + {1}x".format(a, b))  # b가 0보다 크고 c가 0일 때 식 모양
                
                    else:
                        print("입력한 2차 방정식: x^2 + {1}x - {2}".format(a, b, c*(-1))  # b가 0보다 크고 c가 0보다 작을 때 식
                    
                    
                    
                    
                    
                    

            if (b < 0):  # b가 0보다 작을 때 방정식 모양
                if (b == -1)
                    if (c > 0):
                        print("입력한 2차 방정식: x^2 - x + {2}".format(a, b*(-1), c))  # b가 0보다 작고 c가 0보다 클 때 식 모양

                    elif (c == 0):
                        print("입력한 2차 방정식: x^2 - x".format(a, b*(-1)))  # b가 0보다 작고 c가 0일 때 식 모양

                    else:
                        print("입력한 2차 방정식: x^2 - x - {2}".format(a, b*(-1), c*(-1)))  # b가 0보다 작고 c가 0보다 작을 때 식 모양
                
                              
                              
                              
            if (b == 0):  # b가 0일때 방정식 모양
                if (c > 0):
                    print("입력한 2차 방정식: x^2 + {1}".format(a, c))  # b가 0이고 c가 0보다 클 때 식 모양

                elif (c == 0):
                    print("입력한 2차 방정식: x^2".format(a))  # b가 0이고 c가 0일 때 식 모양

                else:
                    print("입력한 2차 방정식: x^2 - {1}".format(a, c*(-1)))  # b가 0이고 c가 0보다 작을 때 식 모양
                    

                    
                    
                
            

    else:#a가 0보다 작을 때 식 모양
        if (b > 0):  # b가 0보다 클 때 방정식 모양
            if (c > 0):
                print("입력한 2차 방정식: - {0}x^2 + {1}x + {2}".format((-1)*a, b, c))  # b가 0보다 크고 c가 0보다 클 때 식 모양
                
            elif (c == 0):
                print("입력한 2차 방정식: - {0}x^2 + {1}x".format((-1)*a, b))  # b가 0보다 크고 c가 0일 때 식 모양
                
            else:
                print("입력한 2차 방정식: - {0}x^2 + {1}x - {2}".format((-1)*a, b, c*(-1)))  # b가 0보다 크고 c가 0보다 작을 때 식 모양

        if (b < 0):  # b가 0보다 작을 때 방정식 모양
            if (c > 0):
                print("입력한 2차 방정식: - {0}x^2 - {1}x + {2}".format((-1)*a, b*(-1), c))  # b가 0보다 작고 c가 0보다 클 때 식 모양
                
            elif (c == 0):
                print("입력한 2차 방정식: - {0}x^2 - {1}x".format((-1)*a, b*(-1)))  # b가 0보다 작고 c가 0일 때 식 모양
                
            else:
                print("입력한 2차 방정식: - {0}x^2 - {1}x - {2}".format((-1)*a, b*(-1), c*(-1)))  # b가 0보다 작고 c가 0보다 작을 때 식 모양

        if (b == 0):  # b가 0일때 방정식 모양
            if (c > 0):
                print("입력한 2차 방정식: - {0}x^2 + {1}".format((-1)*a, c))  # b가 0이고 c가 0보다 클 때 식 모양
                
            elif (c == 0):
                print("입력한 2차 방정식: - {0}x^2".format((-1)*a))  # b가 0이고 c가 0일 때 식 모양
                
            else:
                print("입력한 2차 방정식: - {0}x^2 - {1}".format((-1)*a, c*(-1)))  # b가 0이고 c가 0보다 작을 때 식 모양
    
    D = b ** 2 - 4 * a * c  # 판별식으로 근 개수 구하기
    
    root1 = (-b + (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 첫 번째 근 구하기, 중근이 나올시 이 값 적용할 것
    
    root2 = (-b - (b ** 2 - 4 * a * c) ** (1 / 2)) / (2*a)  # 두 번째 근 구하기
    
    print("---------------------------------------------------------")
    
    if (D > 0):  # 근의 개수와 근 출력, 근은 소수점 6번째 자리 까지 표현
        print("2개의 실수 해가 존재합니다: x = {0:.6f}, x = {1:.6f}".format(root1, root2))  # 판별식이 0보다 클 때 근 2개

    elif (D == 0):
        print("1개의 실수 해가 존재합니다: x = {0:.6f}".format(root1))  # 판별식이 0일 때 근 1개

    else:
        print("실수 해가 존재하지 않습니다")  # 판별식이 0보다 작을 때 근이 존재하지 않음
    
    print("=====================================")
    
else:
    print("---------------------------------------------------------")
    print("계수 a에 0을 입력하지마세요.")
    print("=====================================")








```
