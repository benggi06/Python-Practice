```python
import os
import pickle

def day_check(day):#요일 확인함수
    if day not in '월화수목금':
        return 'error'
         
def nesting_class(day, time, continue_class, time_table_dic):#시간표 중첩 확인함수
    for i in range(time, time + continue_class + 1):
        if i in time_table_dic[day]:
            return 'error'
        else:
            return '0'
                
def time_table(time_table_dic):#시간표 틀
    print('[ TimeTable ]')
    print("-------------")
    print("| |M|T|W|T|F|")
    print('-------------')
    for i in range(1, 9):
        print('|{0}|'.format(i),end='')
        for j in ["월", "화", "수", "목", "금"]:
            if i in time_table_dic[j]:
                print('{0}|'.format(time_table_dic[j][i]),end='')
            else:
                print(' |', end='')
            
        print()
    print('-------------')

def write_table_txt(time_table_dic):
    with open('timetable.txt','w') as file:
        file.write('[ TimeTable ]\n')
        file.write("-------------\n")
        file.write("| |M|T|W|T|F|\n")
        file.write('-------------\n')
        for i in range(1, 9):
            file.write('|{0}|'.format(i))
            for j in ["월", "화", "수", "목", "금"]:
                if i in time_table_dic[j]:
                    file.write('{0}|'.format(time_table_dic[j][i]))
                else:
                    file.write(' |')
            file.write('\n')
        file.write('-------------')
        
def write_table_pkl(time_table_dic):
    with open('timetable.pkl','wb') as file:
        pickle.dump(time_table_dic,file)

def read_table_pkl(time_table_dic):#피클 파일 불러오기
    f = 'timetable.pkl'
    check_file = os.path.isfile(f)#파일이 있는지 확인
    
    if check_file == True:#timetable.pkl파일이 있을때 불러오기
        with open('timetable.pkl','rb') as file:
            time_table_dic = pickle.load(file)
            time_table(time_table_dic)
    else:
        print('파일이 존재하지 않습니다.')
        
```

```python
time_table_dic = {
    '월':{},
    '화':{},
    '수':{},
    '목':{},
    '금':{}
}

miss_choice = False #마지막 선택오류 비활성화

while True:#시작
    
    loadingData = input("저장된 시간표를 불러오시겠습니까? y/n : ")#저장되어있는 시간표 불러오기
    if loadingData == "y":
        read_table_pkl(time_table_dic)
    elif loadingData == "n":
        pass
    else:
        print(">>>[ERROR] y/n 둘 중 하나만 선택하세요!!!")
        continue
        
    while True:#시간표 생성
        if miss_choice == True: #선택오류 활성화시 바로 계속 종료 선택으로 직행
            break

        #요일,시간:그에 관한 예외
        day,time = input('>>> 등록할 수업의 요일과 교시를 입력하세요: ').split()#요일, 시간입력
        if day_check(day) == 'error':
            print('>>>[ERROR] 월~금요일 중을 골라 정확히 입력하세요.')
            break
        time=int(time)#시간변수 타입 정수로 변환
        
        if time > 8 or time < 1:#시간이 8교시를 넘어가거나 0교시보다 작을 때 나오는 오류 메시지
            print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!!!')
            break

        #연강:그에 관한 예외
        continue_class = int(input('>>> 등록할 수업의 연강 시간을 입력하세요: '))#연강시간입력

        if continue_class < 1:#연강시간에 1보다 작은 수를 입력했을 때 오류 메시지
            print('>>> [ERROR] 연강 시간을 정확히 입력하세요!!!')
            break

        elif continue_class+time-1 > 8:#연강+시간이 8을 넘어갈 때 오류 메시지
            print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!!!')
            break
        
        if nesting_class(day, time, continue_class, time_table_dic) == 'error': #시간 중첩 확인
            print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!!!')
            break

        #과목 이름:그에 관한 예외
        sub = input('>>> 등록할 수업의 알파벳을 입력하세요: ')#과목 이름 입력
        
        sub1=sub.upper()#저장된 문자를 대문자로 변환하고 sub1에 저장
        
        alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"#알파벳 저장변수
        
        if len(sub) != 1:#sub의 길이가 1이 아닐때 오류 메시지
            print('>>> [ERROR] 한 개의 대문자 알파벳을 입력하세요!!!')
            break
            
        elif sub1 != sub:#sub이 소문자 알파벳일 때 오류 메시지
            print('>>> [ERROR] 대문자 알파벳을 입력하세요!!!')
            break
        
        elif sub not in alpha:#sub이 알파벳이 아닐때 오류 메시지
            print(">>> [ERROR] 대문자 알파벳을 입력하세요!!!")
            break

        #요일에 관한 예외
        
        for i in range(continue_class):#연강까지 포함해 과목입력 
            time_table_dic[day][time+i] = sub
        
        #만들어진 시간표
        print('>>> 현재 시간표는 다음과 같습니다.\n')
        time_table(time_table_dic)
        
        save_choice = int(input('''(1) 텍스트 저장 (2) 딕셔너리 저장 (3) 둘 다 저장 
    [선택]: '''))
        if save_choice == 1:
            write_table_txt(time_table_dic)
            
        elif save_choice == 2:
            write_table_pkl(time_table_dic)
        elif save_choice == 3:
            write_table_txt(time_table_dic)
            write_table_pkl(time_table_dic)
            
        break
       
    miss_choice = False #선택오류 비활성화
    choice = int(input('''>>> 계속 입력하시겠습니까?:1. 계속 , 2. 종료
    [선택]: ''' ))#종료할건지 계속할건지 선택
    
    if choice == 1:#1: 계속 선택 시
        continue
    
    elif choice == 2:#2: 종료 선택 시
        print('완성된 시간표는 다음과 같습니다.')
        time_table(time_table_dic)
        break
        
    else:#1,2이외를 입력시 선택오류 활성화
        miss_choice = True#활성화
        print('>>> [ERROR] 1과 2를 선택해 제대로 입력하세요!')#오류선택 활성화시 나오는 메시지
        continue
```
