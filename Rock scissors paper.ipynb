```python
import random

li = ["가위","바위","보"]

ch = random.choice(li)

win = 0
draw = 0
lose = 0
while True:
    x = input("가위 바위 보 중 하나를 입력하세요.: ")
#무승부    
    if ch == x:
        print("비겼습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        
        draw +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        
        break
#승리        
    elif (ch == "가위" and x == "바위"):
        print("사용자가 승리했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        
        win +=1
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        
        break
    
    elif (ch == "바위" and x == "보"):
        print("사용자가 승리했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        win +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
            
        break
    
    elif (ch == "보" and x == "가위"):
        print("사용자가 승리했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        win +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        
        break
#패배    
    elif (ch == "가위" and x == "보"):
        print("사용자가 패배했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        lose +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        
        break
        
    elif (ch == "바위" and x == "가위"):
        print("사용자가 패배했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        lose +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        break
        
    elif (ch == "보" and x == "바위"):
        print("사용자가 패배했습니다. 사용자: {0}, 컴퓨터: {1}".format(x,ch))
        rematch = input("다시 하시겠습니까? Y or N: ")
        lose +=1
        
        if rematch == "Y":
            
            continue
            
        print("게임을 종료합니다.")
        break
print("이긴 횟수: {0}, 비긴 횟수: {1}, 진 횟수: {2}".format(win,draw,lose))
```
