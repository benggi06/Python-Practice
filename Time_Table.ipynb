```python
schedule = [[' ']*5 for i in range(8)]#시간표 리스트
miss_choice = False#마지막 선택오류 비활성화

def day_num(day):#요일 숫자 변환 함수
    if day == '월':
        return 0
    elif day == '화':
        return 1
    elif day == '수':
        return 2
    elif day == '목':
        return 3
    elif day == '금':
        return 4

def nesting_class(day_number):#시간표 중첩 확인함수
    for i in range(continue_class):
        if schedule[time+i-1][day_number] ==' ':
            continue
            
        else:
            return 'error'

def day_test(day_number):#요일 확인 및 시간표 채움함수
    if day_number > 4:# day_number가 4보다 클때 오류 생성
        return 'error'
    
    for i in range(continue_class):#연강까지 포함해 과목입력 
        schedule[time+i-1].insert(day_number,sub)

while True:#시작
    
    while True:#시간표 생성
        if miss_choice == True:#선택오류 활성화시 바로 계속 종료 선택으로 직행
            break

        #요일,시간:그에 관한 예외
        day,time = input('>>> 등록할 수업의 요일과 교시를 입력하세요: ').split()#요일, 시간입력
        time=int(time)#시간변수 타입 정수로 변환
        
        if time > 8 or time < 1:#시간이 8교시를 넘어가거나 0교시보다 작을 때 나오는 오류 메시지
            print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!!!')
            break

        #연강:그에 관한 예외
        continue_class = int(input('>>> 등록할 수업의 연강 시간을 입력하세요: '))#연강시간입력

        nesting_class(day_num(day))#시간 중첩 확인

        if nesting_class(day_num(day)) == 'error':#리턴값이 에러로 나올시 오류 메시지
            print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!!!')
            break

        elif continue_class < 1:#연강시간에 1보다 작은 수를 입력했을 때 오류 메시지
            print('>>> [ERROR] 연강 시간을 정확히 입력하세요!!!')
            break

        elif continue_class+time-1 > 8:#연강+시간이 8을 넘어갈 때 오류 메시지
            print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!!!')
            break

        #과목 이름:그에 관한 예외
        sub = input('>>> 등록할 수업의 알파벳을 입력하세요: ')#과목 이름 입력
        
        sub1=sub.upper()#저장된 문자를 대문자로 변환하고 sub1에 저장
        
        alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"#알파벳 저장변수
        
        if len(sub) != 1:#sub의 길이가 1이 아닐때 오류 메시지
            print('>>> [ERROR] 한 개의 대문자 알파벳을 입력하세요!!!')
            break
            
        elif sub1 != sub:#sub이 소문자 알파벳일 때 오류 메시지
            print('>>> [ERROR] 대문자 알파벳을 입력하세요!!!')
            break
        
        elif sub not in alpha:#sub이 알파벳이 아닐때 오류 메시지
            print(">>> [ERROR] 대문자 알파벳을 입력하세요!!!")
            break

        #요일에 관한 예외
        if day_test(day_num(day)) == 'error':#월~금요일이 없을때 오류 메시지
            print('>>>[ERROR] 월~금요일 중을 골라 정확히 입력하세요.')
            break
        #만들어진 시간표
        print('>>> 현재 시간표는 다음과 같습니다.\n')
        print('[ TimeTable ]')
        print("-------------")
        print("| |M|T|W|T|F|")
        print('-------------')
        for i in range(8):#행: 시간
            print('l{0}l'.format(i+1),end='')
            for j in range(5):#열: 날짜
                print('{0}l'.format(schedule[i][j]),end='')
            print()
        print('-------------')
        break
       
    miss_choice = False #선택오류 비활성화
    choice = int(input('''>>> 계속 입력하시겠습니까?:1. 계속 , 2. 종료
    [선택]: ''' ))#종료할건지 계속할건지 선택
    
    if choice == 1:#1: 계속 선택 시
        continue
    
    elif choice == 2:#2: 종료 선택 시
        print('>>> 종료합니다. 완성된 시간표는 다음과 같습니다.\n')#완성된 시간표 보여주기
        print('[ TimeTable ]')
        print("-------------")
        print("| |M|T|W|T|F|")
        print('-------------')
        for i in range(8):
            print('l{0}l'.format(i+1),end='')#행: 시간
            for j in range(5):
                print('{0}l'.format(schedule[i][j]),end='')#열: 날짜
            print()
        print('-------------')
        break
        
    else:#1,2이외를 입력시 선택오류 활성화
        miss_choice = True#활성화
        print('>>> [ERROR] 1과 2를 선택해 제대로 입력하세요!')#오류선택 활성화시 나오는 메시지
        continue
    
```
